```python
Founders=df.groupby('Founders')['Founders'].size().rename('Count').reset_index()
Top5=Founders.sort_values(by='Count',ascending=False).head(5)
plt.pie(Top5['Count'],labels=Top5['Founders'],autopct="%1.2f%%",explode=[0,0.1,0,0,0])
plt.legend(loc='center', bbox_to_anchor=(2.5, 0.5))
plt.title('Top 5 Founders')
plt.show()
```

```python
x= df['Founders'].value_counts().iloc[1:5]
explode = (0, 0.1, 0.05, 0.1)
plt.figure(figsize=(6, 6))
plt.pie(x, labels=x.index, explode=explode, autopct='%1.1f%%', shadow= True, wedgeprops={"width":0.6}, colors=sns.color_palette("twilight_shifted"))
plt.title('Top 4 Founders',fontdict={'fontsize':20}, bbox={'facecolor': 'lightblue'})
plt.show()
```
